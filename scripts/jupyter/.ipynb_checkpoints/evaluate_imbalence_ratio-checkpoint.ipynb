## Densitiy plot of the label to voxel distribution for the paper

```python
#density bandwidth
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html#scipy.stats.gaussian_kde

rc("axes", prop_cycle=cycler(color=plt.cm.Set1.colors))

ratio_coarse = "../../results/l2voxel_ratio_train_weighted_1024_rep.csv"
ratio_fine = "../../results/l2voxel_ratio_train_weighted_1024_dil13_weighted_1024.csv"

ylim = [0, 12]
xlim = [-0.75, 0.75]

organs = ["Parotid_L", "OpticNerve_L"]
dataset = "full"

fig, ax = plt.subplots(ncols=2*len(organs), figsize=(15, 4.5), sharey=True)

df_coarse = pd.read_csv(ratio_coarse)
df_fine = pd.read_csv(ratio_fine)

colors=["C4","C2","C1","C0", "C3"]
labels=[r'$96^{3}$ vx',r'$48^{3}$ vx',r'$24^{3}$ vx',r'$16^{3}$ vx',r' $ 8^{3}$ vx']
#colors=matplotlib.cm.get_cmap('Set2')


for cnt, organ in enumerate(organs):
    df_coarse[organ+"zscore"] = (df_coarse[organ] - df_coarse[organ].mean())/df_coarse[organ].std(ddof=0)
    
    
    for color_cnt, window_size in enumerate([96,48,24,16,8]):
        df_window_size = df_coarse[df_coarse["Windowsize"] == window_size].copy()
        #df_window_size[organ+"zscore"] = (df_window_size[organ] - df_window_size[organ].mean())/df_window_size[organ].std(ddof=0)   
        print(organ, window_size , df_window_size[organ].mean())
        df_window_size[organ+"zscore"].plot.density(ax=ax[cnt], label=labels[color_cnt], bw_method=0.15, linewidth=2, color=colors[color_cnt])
    ax[cnt].legend()
    ax[cnt].set_ylim(ylim)
    ax[cnt].set_xlim(xlim)
    ax[cnt].set_title(organ + " (coarse)")
    ax[cnt].grid(True, axis="y", linestyle='--', linewidth=0.3)
    
for cnt, organ in enumerate(organs):
    df_fine[organ+"zscore"] = (df_fine[organ] - df_fine[organ].mean())/df_fine[organ].std(ddof=0)    
    
    for color_cnt, window_size in enumerate([48,24,16]):
        df_window_size = df_fine[df_fine["Windowsize"] == window_size].copy()
        #df_window_size[organ+"zscore"] = (df_window_size[organ] - df_window_size[organ].mean())/df_window_size[organ].std(ddof=0)   
        print(organ, window_size , df_window_size[organ].mean())
        df_window_size[organ+"zscore"].plot.density(ax=ax[2+cnt], label=labels[1+color_cnt], bw_method=0.15, linewidth=2.2, color=colors[1+color_cnt])
    ax[2+cnt].legend()
    ax[2+cnt].set_ylim(ylim)
    ax[2+cnt].set_xlim(xlim)
    ax[2+cnt].set_title(organ + " (fine)")
    ax[2+cnt].grid(True, axis="y", linestyle='--', linewidth=0.3)
    ax[2+cnt].plot([],[], label="patch", color="white")
    
fig.text(0.5, 0.025, 'z-value', ha='center', fontsize=12)
plt.subplots_adjust(wspace=0.2)
fig.savefig("../../results/result_figures/label2voxelDist.eps", dpi=150, bbox_inches='tight')
```

## AVG Dice + 95 HD to dilation radius plot

```python
file = "../../results/orig_size_res/fine_train_50k.csv"
df = pd.read_csv(file)
df.drop(["File", "Organ"], axis=1, inplace=True)
dl_radii = range(3, 17, 2)

def plotConfig(ax, df, config, labelname, markbest=False): 
    df = df[df['Model'].str.contains(config)]
    df = df[~df['Model'].str.contains("_rep")]
    df = df[~df['Checkpoint'].str.contains("_rep")]

    grouped_df = df.groupby(['Model', 'Checkpoint'])
    df_desc = grouped_df.describe(percentiles=[]).round(4)

    df_desc.reset_index(inplace=True)
    df_desc['dilation'] = 0

    for cnt in dl_radii:
        df_desc.loc[df_desc['Model'].str.contains('_dil'+str(cnt)), 'dilation'] = cnt
    
    df_desc = df_desc.sort_values(by=['dilation'])
    df_desc = df_desc.reset_index(drop=True)
        
    ax[1].plot(df_desc['dilation'], df_desc['95haus_dist','mean'], linewidth=2.2, marker='o', color="C1", label=labelname)
    ax[1].set_title("95HD [mm]")
    ax[1].grid(True, axis="y", linestyle='--', linewidth=0.3)
    ax[1].set_xlabel("Dilation Radius", fontsize=12)
    if markbest:
        ax[1].plot(df_desc['dilation'][5], df_desc['95haus_dist','mean'][5], 'o', markersize=10, color="C0")
    ax[1].legend(prop={'size': 12}, loc='lower left')
    ax[1].xaxis.set_ticks(dl_radii)

    ax[0].plot(df_desc['dilation'], df_desc['dice','mean'], linewidth=2.2, marker='o', color="C1", label=labelname)
    ax[0].set_title("Dice")
    ax[0].grid(True, axis="y", linestyle='--', linewidth=0.3)
    ax[0].set_xlabel("Dilation Radius", fontsize=12)
    if markbest:
        ax[0].plot(df_desc['dilation'][5], df_desc['dice','mean'][5], 'o', markersize=10, color="C0")
    ax[0].legend(prop={'size': 12}, loc='lower left')
    ax[0].xaxis.set_ticks(dl_radii)

rc("axes", prop_cycle=cycler(color=plt.cm.Set1.colors))
fig, ax = plt.subplots(ncols=2, figsize=(15, 1.5))
plt.legend(loc='best')

config = "hr3d_h_e-3_16-72_d_100k__full_e-4_24-24_gdsc_1024s"
labelname = r'$24^{3}$vx - $10^{-4}$ - GDSC'
plotConfig(ax, df.copy(), config, labelname, True)

fig.savefig("../../results/result_figures/dilationRadius.eps", dpi=150, bbox_inches='tight')

```

## Other label to voxel distribution plots to understand the imbalence ratio

```python
def createDensityPlotDataset(resfile, organlist, dataset, window_size_filter=[], ylim=[0, 250], xlim=[0, 1]):
    df = pd.read_csv(resfile)
    df = df[df['Datasize'] == dataset]
   
    fig, ax = plt.subplots(ncols=len(organlist), figsize=(15, 6))        

    for cnt, organ in enumerate(organlist):
        df[organ+"zscore"] = (df[organ] - df[organ].mean())/df[organ].std(ddof=0)

        for window_size in window_size_filter:
            df_window_size = df[df["Windowsize"] == window_size]
            df_window_size[organ+"zscore"].plot.density(ax=ax[cnt], label=window_size, bw_method=0.15)
            ax[cnt].set_ylim(ylim)
            ax[cnt].set_xlim(xlim)
            ax[cnt].set_title(organ+"_"+dataset)
        ax[cnt].legend()
        
        
    #name = dataset
    #fig.get_axes()[0].annotate(name, (0.5, 0.95), xycoords='figure fraction', ha='center')
```

```python
def createHistogram(resfile, organlist, window_size_filter=[], dataset_size_filter=[], ylim=[0, 250], xlim=[0, 1]):
    df = pd.read_csv(resfile)
    df = df[df['Datasize'].isin(dataset_size_filter) & df['Windowsize'].isin(window_size_filter)]
   
    fig, ax = plt.subplots(ncols=len(organlist), figsize=(10, 3))        
    
    for cnt, organ in enumerate(organlist):
        df_no_zero = df[df[organ] > 0.0]
        df_no_zero.hist(column=organ, ax=ax[cnt], bins=50)
        ax[cnt].set_ylim(ylim)
        ax[cnt].set_xlim(xlim)

        
        
    name = '-'.join(str(e) for e in window_size_filter)+"_"+'-'.join(dataset_size_filter)    
    fig.get_axes()[0].annotate(name, (0.5, 0.95), xycoords='figure fraction', ha='center')
    plt.tight_layout()
```
